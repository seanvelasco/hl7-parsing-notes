# Scraping HL7 data

Data was entirely scraped from [Caristix](https://hl7-definition.caristix.com).

API is not exactly public, but the endpoints are exposed when viewing XHR responses under Network Monitor.

```python
base_url = 'https://hl7-definition.caristix.com/v2-api/1/HL7v2.5/Segments/'
```

```python
def fetch(url):
    response = requests.get(url)
    binary_content = response.content
    string_content = (binary_content).decode('utf-8') # Convert to UTF-8 encoding
    filtered = string_content.replace('null', 'None') # 'null' is non-existent in Python, replace with 'None'
    dict_content = eval(filtered) # Convert string into dict/object similar to JSON.parse()
    return dict_content
```

### Scrape all possible segment name

```python
scraped_segments = fetch(base_url)
```

```python
segment_name_list = [item['id'] for item in scraped_segments if len(item['id']) == 3]
print(json.dumps(segment_name_list, indent=4))
```

```python
# [6:] to exclude 'PID - ' from 'PID - Patient Identification'

segment_name_dict = {item['id']: item['label'][6:] for item in scraped_segments if len(item['id']) == 3}
print(json.dumps(segment_name_dict, indent=4))
```

### Scrape characteristics of every segment

```python
def scrape_segment_info(data):
    return [field['name'] for field in data['fields']]
```

```python
# Iterate through each segment in segment_name_list & scrape data about each segment

element_group = {}

for segment_name in segment_name_list: 
    fetched_data = fetch(base_url + segment_name) # https://hl7-definition.caristix.com/v2-api/1/HL7v2.5/Segments/{segment_name}
    scraped_data = scrape_segment_info(fetched_data) 
    element_group[segment_name] = scraped_data # Append data point with segmenet name as key & scraped data as values
```

```python
print(json.dumps(element_group, indent=4))
```

```python
class Parser:

    def __init__(self, char_encoding = ''):
        self.char_encoding = char_encoding
        
    char_encoding = ''
    field_limiter = '|'

    def message_formatter(self, raw):
        try:
            
            ###### FILTER ILLEGAL CHARACTERS ######
            
            # Normalize special characters
            
            message = re.sub(r'\\u00[0-1][A-D|a-d]', '', raw) # Strip start & end MLLP markers
            
            ###### FORM SEGMENTS ######

            message = re.split(r"\\r\\n|\r\n|\\r|\r|\\n|\n", message) # Split single machine string into segments

            message = [segment.replace("\\", '') for segment in message] # Remove backslashes in message
            
            ###### FORM FIELDS ######

            message = [segment.split(self.field_limiter) for segment in message] # Split segments into fields   
            
            ###### EXCLUDE INVALID SEGMENTS ######

            # A valid segment must contain three-character segment name at index 0

            message = [segment for segment in message if len(segment[0]) == 3] # Only include valid segments in message
            
            def msh_exists(): # Check if MSH segment exists in message & is located at index 0 in a segment
                for index, segment in enumerate(message):
                    if (segment[0] == 'MSH'):
                        msh_exists.location = index
                        return True
                        break
                return False
                                
            if msh_exists(): # If MSH exists, make sure message starts with MSH segment & infer encoding characters
                
                if (msh_exists.location != 0):
                    message.insert(0, message.pop(msh_exists.location)) # Move MSH segment to index 0
                    
                delimiter = message[0][1] # Replace default encoding characters
                
                message[0].insert(1, self.field_limiter) # Insert field limiter to MSH segment as a field
                
            # Compile delimiter
                            
            ###### OBJECT CREATION ######

            # Transform segment lists into segment dicts

            message = [{index:
                        {value: splitted for value, splitted in enumerate(component.split('^'), 1)}
                        if re.search(r'\^', component) and component != delimiter # Exclude delimiter in filter
                        else component for index, component in enumerate(position)}
                       for position in message]
            
            # Create list of segment groups that will serve as keys for indexing 
            # Append segment names into dict first as dicts necessitate unique keys, then transform into a list

            segment_groups = list({position[0]: position[0] for position in message})

            # Group segments according segment groups in a dict
            # Append segments into lists first to count the number of segments in a segment group for indexing

            message = {segment_name:
                [segment for segment in message if segment[0] == segment_name] for segment_name in segment_groups}

            # Index segments from 1 to n number of segments in a dict and dicts into segment group dicts
            # [1:] to exclude segment names from every segment to conform to HL7 specs

            message = {segment_name:
                {count: dict(list(segment.items())[1:]) for count, segment in enumerate(message[segment_name], 1)}
                for segment_name in segment_groups}
        
            return message
        except IndexError:
            return ''

            
    def parser(self, message):
        if (isinstance(message, str)):
            message = self.message_formatter(message)
            
        message = {segment_group:
            {"name": segment_name_dict[segment_group]} | 
                {"segments": {segment: {element_group[segment_group][field]: str(message[segment_group][segment][field])
                for field in message[segment_group][segment]
                    if(len(message[segment_group][segment][field]) != 0)}
            for segment in message[segment_group]}}
        for segment_group in message}
        
        return message
    
    def service_field_results(self, raw):
        
        message = raw['OBX']
        
        suffixes = {
            '%': 'PERCENTAGE',
            '#': 'NUMBER',
            'CV|cv': 'COEFFICIENT OF VARIATION',
            'SD|sv': 'STANDARD DEVIATION',
            '[a-zA-Z]*(Histogram)[a-zA-Z]*': 'HISTOGRAM',
        }
    
        def service_field_formatter(field):
        
            suffix_list = []
            for suffix in suffixes: # If a symbol on field matches a record, store symbol on a list to be appended later
                if re.search(suffix, field):
                    suffix_list.append(suffixes[suffix])
                suffix_string = '_'.join(item for item in suffix_list) # Transform list into a single string separated underscores
            
            for diagnostic_class in observation_id:
                
                for name in observation_id[diagnostic_class]: # If field matches a key, replace field with corresponding value from key
                    if re.search(name, field):
                        field = observation_id[diagnostic_class][name]
    
                    service_field_id = '_'.join([field, suffix_string]) # Concatenate field and symbol (eg. PLATELET + COUNT)
                    service_field_id = service_field_id.replace(' ', '_') 
                    service_field_id = service_field_id.rstrip('_' or ' ') 
                    
                try: # Replace below function with type={machine} where machine is HEMATOLOGY, CLINICAL_CHEMISTRY, etc
                    if list(observation_id[diagnostic_class].keys())[list(observation_id[diagnostic_class].values()).index(field)]:
                        service_id = diagnostic_class
                except ValueError:
                    service_id = field
                    
            return {"service_id": service_id, "service_field_id": service_field_id}
                    
            
        def flatten(field):
            if isinstance(field, dict): # Flatten if field is a dict
                field = ' '.join(field[component] for component in field)
            if not isinstance(field, str):
                str(field)
            return field # Always returns a string
    
        results = [{
            "serviceId": service_field_formatter(flatten(message[result][3]))['service_id'],
            "serviceFieldId": service_field_formatter(flatten(message[result][3]))['service_field_id'],
            "quantitativeValue": flatten(message[result][5]),
            "unit": flatten(message[result][6]),
            "refRange": flatten(message[result][7]),
        } for result in message]
    
        return results
```
